#### Task:4 Generate Summary Report from File

```python
# Using only file reading operations:
# 1. Read all sales values from sales_data.txt.
# 2. Convert them into integers.
# 3. Calculate and print:
# - Total sales.
# - Highest Sale.
# - Lowest Sale.
# - Average Sale.
# Do not use any advanced libraries.
with open('sales_data.txt','r') as file:
    lines = file.readlines()
    sales_values = [int(line.strip()) for line in lines]
    total_sales = sum(sales_values)
    highest_sale = max(sales_values)   
    lowest_sale = min(sales_values)
    average_sale = total_sales / len(sales_values) if sales_values else 0
    print(f"Total Sales: {total_sales}")
    print(f"Highest Sale: {highest_sale}")  
    print(f"Lowest Sale: {lowest_sale}")
    print(f"Average Sale: {average_sale}")  
    
```

#### Task 5: Create Product Info File (User Input)

```python
# 1. Ask the user for 3 product names and their prices.
# 2. Write them into a new file products.txt in this format:
# 3. ProductName | Price
# 4. Read the file and print each line with proper formatting.

product = []
for i in range(1,4):
    product_name=input("Enter product name :")
    price=int(input("Enter the price :"))
    product.append((product_name, price))
with open('products.txt','w') as prod_file:    
        for name, price in product:
            prod_file.write(f"{name} | {price}\n")
with open('products.txt','r') as prod_file:
    lines = prod_file.readlines()
    for line in lines:
        name, price = line.strip().split(' | ')
        print(f"Product: {name}, Price: {price}")
        

```

#### Task 7: Mini Project - Export Discounted Prices


```python
#create a dictionary:
prices = {
    "Mouse": 500,
    "Keyboard": 800,
    "Monitor": 7000,
    "Pendrive": 400,
    "Camera": 5000
}
#Ask the user for a discount percentage.
#Write discounted prices into discount_report.txt using:
#Product | Original Price | Discounted Price
#After writing, read the file and print it to the terminal.

try:
    discount_percent = float(input("Enter discount percentage: "))
except ValueError:
    print("Invalid input! Please enter a number.")
    exit()

# Calculate discounted prices
discounted_prices = {}
for product, price in prices.items():
    discounted_prices[product] = price - (price * discount_percent / 100)

# Write to file
with open("discount_report.txt", "w") as f:
    f.write("Product  | Original Price | Discounted Price\n")
    f.write("-" * 45 + "\n")
    for product, price in prices.items():
        f.write(f"{product:8} | {price:14} | {discounted_prices[product]:.2f}\n")

# Read and print the file
with open("discount_report.txt", "r") as f:
    report_content = f.read()
    print(report_content)
```

```python
#Extra (Optional): Write a summary at the bottom of the file:
#Total Items:X
#Average Discounted Price: Y

total_items = len(prices)
avg_discounted_price = sum(discounted_prices.values()) / total_items
with open("discount_report.txt", "a") as f:
    f.write("\nSummary:\n")
    f.write(f"Total Items: {total_items}\n")
    f.write(f"Average Discounted Price: {avg_discounted_price:.2f}\n")
# Read and print file content
with open("discount_report.txt", "r") as f:
    content = f.read()
    print(content)
```
