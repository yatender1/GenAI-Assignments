#### Task 1 - Basic Function: Price After Discount

```python
# Write a function apply_discount(price, discount_percent) that:
#1. Returns the price after discount
#2. If discount_percent is missing, apply a default discount of 5%
#3. Test the function with:
#4. apply_discount(1000,10)
#5. apply_discount(500)# uses default discount
# Extra (optional): Add a condition inside the function to ensure discount never exceeds 60%

def apply_discount(price,discount_percent=5):
    if discount_percent>60:
        discount_percent=60

    final_price=price-(price*discount_percent)/100
    return final_price

print(f"Apply discount (1000,10) = ", apply_discount(1000,10))
print(f"Apply_discount(500)) = ", apply_discount(500))
print(f"Apply_discount(1000,90) = ", apply_discount(1000,90))
```

#### Task 2 - Recursive Function: Factorial Utility

```python
#Create a recursive function factorial(n) that:
#1. Returns the factorial of n.
#2. Handles two edge cases: n==0 and n==1.
#3. Prints an error message if n is negative.
#Test with:
#factorial(5)
#factorial(0)
#factorial(-3)

def factorial_recursive(n):
    if n>1:
        return n*factorial_recursive(n-1)
    elif n==0 or n==1:
        return 1
    else:
        print("Error, Factorial is not defined for negative numbers")
        return None

print("factorial(5) =", factorial_recursive(5))  
print("factorial(0) =", factorial_recursive(0))   
print("factorial(-3) =", factorial_recursive(-3)) 
```

#### Task 3: Lambda Function: GST Calculator

```python
#Create a lambda function gst that returns price after adding 18% GST
#Example:
gst=lambda price:price+(0.18*price)
print(gst(100))# should return 118
#Extra(optional): Create another lambda to compute final price after GST + discount together

Final_price= lambda price, discount: (price +(0.18*price))-((price+(0.18*price))*discount/100)

print("Final Price(1000,10) =", Final_price(1000,10))
print("Final Price(500,5) =", Final_price(500,5))
```

#### Task 4- Using map(): Apply GST to List of Prices

```python
#Given:
#prices=[100,250,400,1200,50]
#Use map() with your gst lambda to generate a new list prices_with_gst. 
#Print both lists:
#Original prices
#Prices after GST

prices=[100,250,400,1200,50]
gst=lambda x:x*0.18
prices_with_gst=list(map(gst,prices))
print("Original prices:",prices)
print("Prices after GST:",prices_with_gst)
```

#### Task 5 - Using filter(): Fileter Expensive Products


```python
#Given the list:
prices=[100,250,400,1200,50,2000,850]
#use filter() to:
#1. create a list of prices greater than 500.
#2. create another list of prices less than or equal to 500.
#Print both lists.
greater_prices=list(filter(lambda x:x>500,prices))
lesser_prices=list(filter(lambda x:x<=500,prices))
print(f"Prices greater than 500: {greater_prices}")
print(f"Pricess less than or equal to 500: {lesser_prices}")       
```

#### Task 6- Combined Utility Function

```python
# Write a function process_prices that:
# 1. Takes a list of prices.
# 2. Uses map() + lambda to apply 10% discount to all prices.
# 3. Uses filter() to keep only discounted prices that are above 300.
# 4. Returns both lists:
#   - discounted_prices
#   - filtered_prices
#   Test with:
#   process_prices([100,500,900,50,750])  
def process_prices(prices):
    discounted_prices=list(map(lambda x:x-(x*0.10),prices))
    filtered_prices=list(filter(lambda x:x>300,discounted_prices))
    return discounted_prices,filtered_prices
print(process_prices([100,500,900,50,750]))
```

```python
#### Task 7 - Mini Problem: Menu Using Functions
# Create the three small functions:
# 1. add_price(price_list, price) -> adds price to the list.
# 2. get_average_price(price_list) -> returns the average price.
# 3. get_max_price(price_list) -> returns the maximum price.
#    Then create a simple menu:
# 1 -> Add price
# 2 -> Show average price
# 3 -> Show highest price
# q -> Quit
#   Use only loops + functions call (no OOP).
def add_price(price_list, price):
    price_list.append(price)
    print(f"{price} added successfully")

def get_average_price(price_list):
    if len(price_list)==0:
        print("No prices available to calculate average")
        return None
    get_average=sum(price_list)/len(price_list)
    return get_average

def get_max_price(price_list):
    if len(price_list)==0:
        print("No prices available to calculate maximum price")
        return None
    max_price=max(price_list)
    return max_price

price_list=[]

while True:
    print("/Menu")
    print("1. Add price")
    print("2. Show average price")
    print("3. Show highest price")
    print("q - Quit")

    choice=input("Enter your choice :")

    if choice == "1":
        try:
            price=int(input("Enter price to add : "))
            add_price(price_list,price)
        except ValueError:
            print("Invalid price, Enter numeric value")
    elif choice == "2":
        average=get_average_price(price_list)
        if average is not None:
             print(f"Average Price: {average}")
    elif choice=="3":
        max_price=get_max_price(price_list)
        if max_price is not None:
            print(f"Highest Price: {max_price}")
    elif choice.lower()=='q':
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please select a valid option from the menu.")
```
