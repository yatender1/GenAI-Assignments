```python
# 2. Apply the following discount rules and print the final amount:
# order_amount >=2000 -> 15% discount
# 1500 <=order_amount <2000 -> 10% discount
# 1000 <= order_amount <1500 -> 7% discount
# otherwise -> 0% discount
order_amount=int(input("Enter order amount :"))
if order_amount >= 2000:
    final_amount=order_amount-((order_amount*15)/100)
    print(f"{order_amount} is order amount and final amount after 15% discount is : {final_amount}")
elif order_amount >= 1500:
    final_amount=order_amount-((order_amount*10)/100)
    print(f"{order_amount} is order amount and final amount after 10% discount is : {final_amount}")
elif order_amount >=1000:
    final_amount=order_amount-((order_amount*7)/100)
    print(f"{order_amount} is order amount and final amount after 7% discount is : {final_amount}")
else:
    final_amount=order_amount
    print(f"{order_amount} is order amount and final amount after 0% discount is : {final_amount}")
```

#### Task2: Process Multiple Orders (for loop)

```python
#1. Given a list of order amount: orders= [1200,2500,800,1750,3000], use a for loop to apply the discount rules from Task 1 to each order and print a summary table showing: order_amount ->discount% ->final_amount

orders=[1200,2500,800,1750,3000]
print("Order Amount | Discount % | Final Amount")
print("-----------------------------------------")
for l in orders:
    if l>=2000:
        discount=0.15
    elif l>=1500:
        discount=0.10
    elif l>=1000:
        discount=0.07
    else:
        discount=0.0
    
    final_amount = l-(l*discount)
    print(f"{l:<13}|{discount*100:<12.0f}|{final_amount:<1.0f}")
    
```

```python
# 2. Also compute the print the total revenue after discounts

orders=[1200,2500,800,1750,3000]
Total_Revenue=0
print("Order Amount | Discount % | Final Amount | Total Revenue")
print("--------------------------------------------------------")
for l in orders:
    if l>=2000:
        discount=0.15
    elif l>=1500:
        discount=0.10
    elif l>=1000:
        discount=0.07
    else:
        discount=0.0
    
    final_amount = l-(l*discount)
    Total_Revenue= Total_Revenue+final_amount
    print(f"{l:<13}|{discount*100:<12.0f}|{final_amount:<14.0f}|{Total_Revenue:<.0f}")
print("--------------------------------------------------------")
print(f"Total Revenue after discounts: {Total_Revenue:.2f}")
```

```python
# Extra (Optional): Print the number of orders that received a discount (discount>0)

orders1=[1200,2500,800,1750,3000]
Total_Revenue=0
discounted_orders=0
print("Order Amount | Discount % | Final Amount | Total Revenue")
print("--------------------------------------------------------")
for l in orders1:
    if l>=2000:
        discount=0.15
    elif l>=1500:
        discount=0.10
    elif l>=1000:
        discount=0.07
    else:
        discount=0.0
    
    final_amount = l-(l*discount)
    Total_Revenue= Total_Revenue+final_amount
    if discount>0:
        discounted_orders += 1
    print(f"{l:<13}|{discount*100:<12.0f}|{final_amount:<14.0f}|{Total_Revenue:<.0f}")
print("--------------------------------------------------------")
print(f"Total Revenue after discounts: {Total_Revenue:.2f}")
print(f"Total number of discounted orders: {discounted_orders}")

```

#### Task 3 User Menu(while loop + break/continue)

```python
# 1. Implement a simple while-loop drien menu that keeps asking the user to choose an action until they press q to quit.
# Menu options:
# 1. Add order amount to a running list(use(input))
# 2. Show all orders and totals afer applying discounts
# q-quit

myorder = []
while True:
    print("\nMenu:")
    print("1. Add order amount: ")
    print("2. Show all orders with discounts and totals")
    print("q. - Quit")

    choice=input("Enter your choice: ").strip().lower()
    
    if choice == "1":
        try:
            amount_order=int(input("Enter Order Amount:"))
            myorder.append(amount_order)
            print("Order added successfully")
        except ValueError:
            print("Error, Enter numeric value")
    
    elif choice == "2":
        if not myorder:
            print("No orders yet.")
        else:
            print("\nOrder Amount | Discount % | Final Amount")
            print("-----------------------------------------")

            revenue=0
            discount_orders=0

            for amount_order in myorder:
                if amount_order>=2000:
                    d_discount=0.15
                elif amount_order>=1500:
                    d_discount=0.10
                elif amount_order>=1000:
                    d_discount=0.07
                else:
                    d_discount=0.0

                f_amount=amount_order-(amount_order*d_discount)
                revenue+=f_amount

                if d_discount>0:
                    discount_orders+=1

                print(f"{amount_order:<12} | {d_discount*100:<10.0f} | {f_amount:<12.2f}")

            print("-----------------------------------------")
            print(f"Total Revenue after discounts: {revenue:.2f}")
            print(f"Number of orders with discount: {discount_orders}")
        
    elif choice == "q":
        print("Exiting program")
        break

    else:
        print("Invalid choice. Please try again.")
        
```

```python
# 2. Use continue to re-show the menu after invalid input and break to exit on q.
# Constraints: Keep the loop logic simple and use only lists and loops(no function required)


myorder = [] 

while True:
    print("\nMenu:")
    print("1. Add order amount")
    print("2. Show all orders with discounts and totals")
    print("q. Quit")

    choice = input("Enter your choice: ").strip().lower()

    if choice == "1":
        try:
            amount_order = int(input("Enter Order Amount: "))
            myorder.append(amount_order)
            print("Order added successfully!")
        except ValueError:
            print("Error: Please enter a valid numeric order amount.")
        continue  

    elif choice == "2":
        if not myorder:
            print("No orders yet.")
        else:
            print("\nOrder Amount | Discount % | Final Amount")
            print("-----------------------------------------")

            revenue = 0
            discount_orders = 0

            for amount_order in myorder:
                if amount_order >= 2000:
                    d_discount = 0.15
                elif amount_order >= 1500:
                    d_discount = 0.10
                elif amount_order >= 1000:
                    d_discount = 0.07
                else:
                    d_discount = 0.0

                f_amount = amount_order - (amount_order * d_discount)
                revenue += f_amount

                if d_discount > 0:
                    discount_orders += 1

                print(f"{amount_order:<12} | {d_discount*100:<10.0f} | {f_amount:<12.2f}")

            print("-----------------------------------------")
            print(f"Total Revenue after discounts: {revenue:.2f}")
            print(f"Number of orders with discount: {discount_orders}")
        continue  

    elif choice == "q":
        print("Exiting program.")
        break  

    else:
        print("Invalid choice. Please try again.")
        continue  
```

```python
#Given a list of daily sales: daily =[200,150,0,400,50,-1,300], iterate through it with a for loop and:
#- If a day's value is -1, treat it as corrupted data and break the loop(stop processing).
#- If a day's value is 0, treat it as a day with no sales and continue (skip adding to revenue).
#- for valid positive sales, add to total_sales and print the running total.

daily=[200,150,0,400,50,-1,300]
total_sales=0
for value in daily:
    if value==-1:
        print("Corrupted Data, Stop processing")
        break
    elif value==0:
        print("Day with no sales, Skipping")
    else:
        total_sales+=value
        print(f"Total Sales {total_sales}")

print(f"\nTotal running sales is {total_sales}")
```
