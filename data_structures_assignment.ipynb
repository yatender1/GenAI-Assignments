```python
# Extra (Optional): Print the product with both the maximum and minium prices
max_product=max(price_dict,key=price_dict.get)
max_price=price_dict[max_product]
min_product=min(price_dict,key=price_dict.get)
min_price=price_dict[min_product]
print("Product with maximum price :", max_product, "-", max_price)
print("Prodcut with minumum price :",min_product,"-",min_price)
```

# Task4: Combined Operations

```python
# Using the products list and price_dict, create a list of tuples named catalog where each tuple is (product_name, price, category)

catalog = []
for product, category in zip(products, categories):
    price = price_dict.get(product, None)  # handle missing prices safely
    catalog.append((product, price, category))

print(catalog)
```

```python
# From catalog, create a new dictionary category_to_products that maps each category to a list or products in that category.

category_to_products = {}

for product, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(product)

print(category_to_products)
```

```python
# Print all the products that belong to the category that has maximum number of Prodcuts.
category_to_products = {}
for product, price, category in catalog:
    category_to_products.setdefault(category, []).append(product)

# Find category with maximum number of products
max_category = max(category_to_products, key=lambda c: len(category_to_products[c]))

# Print products in that category
print("Category with maximum products:", max_category)
print("Products:", category_to_products[max_category])

```
